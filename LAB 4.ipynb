```python
import threading

def worker(tid, barrier):
    # Phase 1 computation
    print(f"Thread {tid}: Performing Phase 1 computation")

    # Barrier synchronization
    barrier.wait()

    # Phase 2 computation
    print(f"Thread {tid}: Performing Phase 2 computation")


num_threads = 4
barrier = threading.Barrier(num_threads)
threads = []

for i in range(num_threads):
    t = threading.Thread(target=worker, args=(i, barrier))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("All threads have completed both phases.")
```

```python
import threading

# Total size of array
N = 8

# Shared arrays
A = [0] * N
B = [0] * N

# Initialize array A
for i in range(N):
    A[i] = i + 1

# Number of threads
num_threads = 4

# Barrier synchronization object
barrier = threading.Barrier(num_threads)

# Worker function
def worker(thread_id, start, end):

    # -------- Phase 1 --------
    for i in range(start, end):
        B[i] = A[i] * 2
        print(f"Thread {thread_id} completed Phase 1 for index {i}")

    # -------- Barrier Synchronization --------
    barrier.wait()

    # -------- Phase 2 --------
    for i in range(start, end):
        B[i] = B[i] + 5
        print(f"Thread {thread_id} completed Phase 2 for index {i}")


# Creating threads
threads = []
chunk_size = N // num_threads

for t in range(num_threads):
    start = t * chunk_size
    end = (t + 1) * chunk_size

    thread = threading.Thread(target=worker, args=(t, start, end))
    threads.append(thread)
    thread.start()

# Wait for all threads to finish
for thread in threads:
    thread.join()

# Final Output
print("\nFinal Result Array B:")
print(B)

print("\nTwo-phase computation completed successfully.")
```

```python
import threading

N = 6

# Initialize image and output
image = [[i * N + j + 1 for j in range(N)] for i in range(N)]
output = [[0 for _ in range(N)] for _ in range(N)]

num_threads = 4
barrier = threading.Barrier(num_threads)

def worker(tid):
    # -------- Phase 1: Normalization --------
    for i in range(tid, N, num_threads):
        for j in range(N):
            image[i][j] = image[i][j] // 2

    # -------- Barrier Synchronization --------
    barrier.wait()

    # -------- Phase 2: 3Ã—3 Convolution --------
    for i in range(1 + tid, N - 1, num_threads):
        for j in range(1, N - 1):
            total = (
                image[i-1][j-1] + image[i-1][j] + image[i-1][j+1] +
                image[i][j-1]   + image[i][j]   + image[i][j+1] +
                image[i+1][j-1] + image[i+1][j] + image[i+1][j+1]
            )
            output[i][j] = total // 9  # Averaging filter


# Create and start threads
threads = []
for t in range(num_threads):
    thread = threading.Thread(target=worker, args=(t,))
    threads.append(thread)
    thread.start()

# Wait for all threads
for thread in threads:
    thread.join()

# Display output
print("\nFiltered Image:")
for i in range(1, N - 1):
    for j in range(1, N - 1):
        print(f"{output[i][j]:3d}", end=" ")
    print()
```
